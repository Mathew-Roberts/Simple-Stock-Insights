```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

# Reading in the Crude Oil Data

```python
# READING IN THE DATA SET
OilDF = pd.read_csv('CL=F_oil.csv')


#setting Date as index
OilDF['Date'] = pd.to_datetime(OilDF['Date'])
OilDF.index = OilDF['Date']

#Finding Nan values and using zero order interpolation to hold previous day's value

cols = OilDF.columns
if OilDF.isnull().values.any() == True:
    for col in cols:
        OilDF[col] = OilDF[col].interpolate(method = 'linear')

#Ensuring numerical columns are float64 objects and dropping date column
OilDF = OilDF.drop(columns = 'Date')

float_cols = OilDF.columns[0:]
for col in float_cols:
    OilDF[col] = pd.to_numeric(OilDF[col])
 
    

print(OilDF.head())
print(OilDF.dtypes)
OilDF.isnull().values.any()
len(OilDF)
        
```

# Reading in the Natural Gas Data

```python
#REading in the second data set
GasDF = pd.read_csv('NG=F_gas.csv')
GasDF.head()

#setting Date as index
GasDF['Date'] = pd.to_datetime(GasDF['Date'])
GasDF.index = GasDF['Date']



#Finding Nan values and using zero order interpolation to hold previous day's value

cols = GasDF.columns
if GasDF.isnull().values.any() == True:
    for col in cols:
        GasDF[col] = GasDF[col].interpolate(method = 'linear')
    

#Ensuring numerical columns are float64 objects and ensuring the date column is dropped
GasDF = GasDF.drop(columns = 'Date')

float_cols = GasDF.columns[0:]
for col in float_cols:
    GasDF[col] = pd.to_numeric(GasDF[col])
    
print(GasDF.dtypes)
print(GasDF.head()) 
GasDF.isnull().values.any()

```

# Reading in the Exxon Mobil Corp Data 

```python
#Reading in the Oil and Gas Investments data set
InvDF = pd.read_csv('XOM.csv')
InvDF.head()

#setting Date as index
InvDF['Date'] = pd.to_datetime(InvDF['Date'])
InvDF.index = InvDF['Date']

#Finding Nan values and using zero order interpolation to hold previous day's value

cols = InvDF.columns
if InvDF.isnull().values.any() == True:
    for col in cols:
        InvDF[col] = InvDF[col].interpolate(method = 'linear')
        

#Ensuring numerical columns are float64 objects and ensuring the date column is dropped
InvDF = InvDF.drop(columns = 'Date')

float_cols = InvDF.columns[0:]
for col in float_cols:
    InvDF[col] = pd.to_numeric(InvDF[col])        
        

print(InvDF.head())
print(InvDF.dtypes)
InvDF.isnull().values.any()
```

# Oil Data 

```python
### Lets see the open and close between 2015 and 2020

for col in OilDF.columns:
    if col == 'Open':
        plt.plot(OilDF.index, OilDF[col], linewidth = 1,color = 'r', label = 'Open Share Price')
  

plt.xlabel('Date',size = 15)
plt.ylabel('Price per Barrel (USD)',size= 15)
plt.gcf().autofmt_xdate()
plt.title('2015-2020 Crude Oil Open Share Prices',size =15)
plt.legend()
```

```python
# Lets look at the yearly averages 'opening share prices'
OilAv16 = OilDF['Open'].loc['21/12/2015':'21/12/2016'].mean()
OilStd16 =(OilDF['Open'].loc['21/12/2015':'21/12/2016'].var())**0.5

OilAv17 = OilDF['Open'].loc['21/12/2016':'21/12/2017'].mean()
OilStd17 =(OilDF['Open'].loc['21/12/2016':'21/12/2017'].var())**0.5

OilAv18 = OilDF['Open'].loc['21/12/2017':'21/12/2018'].mean()
OilStd18 =(OilDF['Open'].loc['21/12/2017':'21/12/2018'].var())**0.5

OilAv19 = OilDF['Open'].loc['21/12/2018':'21/12/2019'].mean()
OilStd19 =(OilDF['Open'].loc['21/12/2018':'21/12/2019'].var())**0.5

OilAv20 = OilDF['Open'].loc['21/12/2019':].mean()
OilStd20 =(OilDF['Open'].loc['21/12/2019':].var())**0.5


#Plotting a bar chart
xvals = np.arange(2016,2021,1)
OilAverages = [OilAv16,OilAv17,OilAv18,OilAv19,OilAv20]
OilStds = [OilStd16,OilStd17,OilStd18,OilStd19,OilStd20]
plt.bar(xvals,OilAverages,color = 'b',label = 'Average Oil Price')
plt.bar(xvals,OilStds, color = 'r', label = 'Standard Deviation')
plt.xlabel('Date')
plt.ylabel('Average Price per Barrel (USD)')
plt.title('2016-2020 Average Price of Crude Oil')
plt.legend()
```

```python
#Oil rolling average
#Lets put in a rolling average
OilDF['Volume Rolling Average'] = OilDF['Volume'].rolling(window=100).mean()



#Lets see what the volume data looks like (volume of trades)
plt.plot(OilDF.index,OilDF['Volume'],color = 'g')
plt.plot(OilDF.index,OilDF['Volume Rolling Average'],color = 'r')
plt.gcf().autofmt_xdate()
plt.xlabel('Date')
plt.ylabel('Volume of Trades (10 million)')
plt.title('2015-2020 Volume of Trades')
```
